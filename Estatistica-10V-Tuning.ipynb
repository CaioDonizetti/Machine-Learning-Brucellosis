### **RNA**

```python
#---------------------------------- RNA_rep -----------------------------------------------------
combinacoes = np.array(RNA_rep['0'].unique())
avaliacoes = [['Hiperparâmetros','Sens. Méd.','Sens. DP', 'Esp. Méd.','Esp. DP', 'IBA Méd.','IBA DP' ]]
for hiperparametros in combinacoes:
    combinacao = hiperparametros.replace("{'activation':", "Func. Ativ. =")
    combinacao = combinacao.replace("'hidden_layer_sizes':", "Nº Neur. =")
    combinacao = combinacao.replace("}", "")
    combinacao = combinacao.replace("'logistic'", "Sig.")
    combinacao = combinacao.replace("'tanh'", "Tan. Hip.")
    combinacao = combinacao.replace("'relu'", "Relu")
    avaliacoes.append([combinacao, 
                '{:.2f}'.format(RNA_rep[RNA_rep['0'].str.contains(hiperparametros)]['1'].mean()*100,2),
                '{:.2f}'.format(RNA_rep[RNA_rep['0'].str.contains(hiperparametros)]['1'].std()*100,2),
                '{:.2f}'.format(RNA_rep[RNA_rep['0'].str.contains(hiperparametros)]['2'].mean()*100,2),
                '{:.2f}'.format(RNA_rep[RNA_rep['0'].str.contains(hiperparametros)]['2'].std()*100,2),
                '{:.2f}'.format(RNA_rep[RNA_rep['0'].str.contains(hiperparametros)]['6'].mean()*100,2),
                '{:.2f}'.format(RNA_rep[RNA_rep['0'].str.contains(hiperparametros)]['6'].std()*100,2)])

tabela = pd.DataFrame(data=avaliacoes[1:], columns=avaliacoes[0]).sort_values('IBA Méd.', ascending=False)
tabela.to_csv("Estatistica-Tuning/RNA_10V_rep.csv", encoding = 'utf-8', index = False)


#---------------------------------- RNA_adasyn -----------------------------------------------------
combinacoes = np.array(RNA_adasyn['0'].unique())
avaliacoes = [['Hiperparâmetros','Sens. Méd.','Sens. DP', 'Esp. Méd.','Esp. DP', 'IBA Méd.','IBA DP' ]]
for hiperparametros in combinacoes:
    combinacao = hiperparametros.replace("{'activation':", "Func. Ativ. =")
    combinacao = combinacao.replace("'hidden_layer_sizes':", "Nº Neur. =")
    combinacao = combinacao.replace("}", "")
    combinacao = combinacao.replace("'logistic'", "Sig.")
    combinacao = combinacao.replace("'tanh'", "Tan. Hip.")
    combinacao = combinacao.replace("'relu'", "Relu")
    avaliacoes.append([combinacao, 
                '{:.2f}'.format(RNA_adasyn[RNA_adasyn['0'].str.contains(hiperparametros)]['1'].mean()*100,2),
                '{:.2f}'.format(RNA_adasyn[RNA_adasyn['0'].str.contains(hiperparametros)]['1'].std()*100,2),
                '{:.2f}'.format(RNA_adasyn[RNA_adasyn['0'].str.contains(hiperparametros)]['2'].mean()*100,2),
                '{:.2f}'.format(RNA_adasyn[RNA_adasyn['0'].str.contains(hiperparametros)]['2'].std()*100,2),
                '{:.2f}'.format(RNA_adasyn[RNA_adasyn['0'].str.contains(hiperparametros)]['6'].mean()*100,2),
                '{:.2f}'.format(RNA_adasyn[RNA_adasyn['0'].str.contains(hiperparametros)]['6'].std()*100,2)])

tabela = pd.DataFrame(data=avaliacoes[1:], columns=avaliacoes[0]).sort_values('IBA Méd.', ascending=False)
tabela.to_csv("Estatistica-Tuning/RNA_10V_adasyn.csv", encoding = 'utf-8', index = False)


#---------------------------------- RNA_contr -----------------------------------------------------
combinacoes = np.array(RNA_contr['0'].unique())
avaliacoes = [['Hiperparâmetros','Sens. Méd.','Sens. DP', 'Esp. Méd.','Esp. DP', 'IBA Méd.','IBA DP' ]]
for hiperparametros in combinacoes:
    combinacao = hiperparametros.replace("{'activation':", "Func. Ativ. =")
    combinacao = combinacao.replace("'hidden_layer_sizes':", "Nº Neur. =")
    combinacao = combinacao.replace("}", "")
    combinacao = combinacao.replace("'logistic'", "Sig.")
    combinacao = combinacao.replace("'tanh'", "Tan. Hip.")
    combinacao = combinacao.replace("'relu'", "Relu")
    avaliacoes.append([combinacao, 
                '{:.2f}'.format(RNA_contr[RNA_contr['0'].str.contains(hiperparametros)]['1'].mean()*100,2),
                '{:.2f}'.format(RNA_contr[RNA_contr['0'].str.contains(hiperparametros)]['1'].std()*100,2),
                '{:.2f}'.format(RNA_contr[RNA_contr['0'].str.contains(hiperparametros)]['2'].mean()*100,2),
                '{:.2f}'.format(RNA_contr[RNA_contr['0'].str.contains(hiperparametros)]['2'].std()*100,2),
                '{:.2f}'.format(RNA_contr[RNA_contr['0'].str.contains(hiperparametros)]['6'].mean()*100,2),
                '{:.2f}'.format(RNA_contr[RNA_contr['0'].str.contains(hiperparametros)]['6'].std()*100,2)])

tabela = pd.DataFrame(data=avaliacoes[1:], columns=avaliacoes[0]).sort_values('IBA Méd.', ascending=False)
tabela.to_csv("Estatistica-Tuning/RNA_10V_contr.csv", encoding = 'utf-8', index = False)

```
